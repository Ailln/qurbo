# MIQP 数据读取示例

本 notebook 演示如何读取本文件夹下的 `.npz` 数据文件，并查看每个字段的维度。

```python
%load_ext autoreload
%autoreload 2
```

```python
from pathlib import Path
import numpy as np

DATA_DIR = Path.cwd()
files = sorted(DATA_DIR.glob('*.npz'))
files
```

```python
def load_miqp_data(path):
    data = np.load(path, allow_pickle=True)
    return {key: data[key] for key in data.files}

instance = load_miqp_data(files[0])
print(files[0].name)
for key, value in instance.items():
    print(f'{key:14s} shape={value.shape}, dtype={value.dtype}')
```

```python
# 常用字段示例
n = int(instance['n'])
p = int(instance['p'])
Q = instance['Q']
c = instance['c']
A = instance['A']
G = instance['G']
B = instance['B']

print('n =', n, 'p =', p)
print('Q shape =', Q.shape)
print('c shape =', c.shape)
print('A shape =', A.shape)
print('G shape =', G.shape)
print('B shape =', B.shape)
```

```python
# 批量读取全部实例
all_instances = {path.stem: load_miqp_data(path) for path in files}
for name, item in all_instances.items():
    print(name, 'n=', int(item['n']), 'p=', int(item['p']), 'm1=', int(item['m1']), 'm2=', int(item['m2']))
```
